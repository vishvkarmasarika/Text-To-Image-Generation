```python
from diffusers import StableDiffusionPipeline
import torch
from google.colab import userdata
import os

# Ensure the HF_TOKEN environment variable is set
# os.environ['HF_TOKEN'] = userdata.get('HF_TOKEN') # Removed this line

# Load the pipeline, explicitly passing the token
pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    torch_dtype=torch.float16,
    use_safetensors=True,
    use_auth_token=os.environ['HF_TOKEN'] # Explicitly pass the token
).to("cuda")
```

```python
import matplotlib.pyplot as plt
user_prompt = input("enter your image prompt:")
image= pipe(user_prompt). images[0]
plt.imshow(image)
plt.axis("off")
plt.title("generated image")
plt.show()
```
